```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Zomato_Data = pd.read_csv('zomato.csv')
```

```python
df = df.drop(["url","address","phone","reviews_list","menu_item","listed_in(city)","listed_in(type)"],axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.rename(columns={'approx_cost(for two people)': 'CostForTwo'}, inplace=True)
```

```python
df
```

```python
df['rate'].unique()
```

```python
def replace_rate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value=str(value).split("/")
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(replace_rate)
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

## Filling null value

```python
df["rate"].fillna(df["rate"].mean(),inplace=True)
df["rate"].isnull().sum()
```

```python
df.info()
```

```python
df["location"].unique()
```

```python
df['CostForTwo'].unique()
```

```python
df["location"].value_counts()
```

```python
location=df["location"].value_counts(ascending=False)
```

```python
location_lessthan300=location[location<300]

def new_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df["location"]=df["location"].apply(new_location)
df["location"].value_counts()
```
